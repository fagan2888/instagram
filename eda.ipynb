```python
import pandas as pd
# load mass dataframe
df=pd.read_csv('data/scraped/All_users_ttv.princearthur_20190613_2119.csv')
# tell us about it
print(df.info())
# give us a sample
df.sample(7)
```

```python
# dataframe of followers
follows_princearthur=[url for url in df.loc[df.user_followed_by==True].user_profile]
# dataframe of accounts we're following
princearthur_follows=[url for url in df.loc[df.user_follows==True].user_profile]
# dataframe of accounts we're following that do not follow us
eligible_for_unfollowing=[url for url in princearthur_follows if url not in follows_princearthur]
```
